```python
import datetime
import json
import logging
import requests
import tkinter as tk
import urllib
from typing import List, Dict, Any
from classes import Settings
config = Settings.WEATHER
```

```python
params: Dict[str, str] = {
    "lat": config["lat"],
    "lon": config["lon"],
    "exclude": config["exclude"],
    "appid": config["apiKey"],
    "units": config["units"]
}
query: str = config["baseUrl"] + urllib.parse.urlencode(params)

# call API
r = requests.get(query)
weather = json.loads(r.text)
currentWeather: Dict[str, Any] = {k: weather["current"][k] for k in config["desiredFields"]}
```

```python
output = currentWeather
config["imgUrl"].format(imgCode=output["weather"][0]["icon"])
```
