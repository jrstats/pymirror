```python
import datetime
import time
from crontab import CronTab

cronSyntax = "* * * * * * *"
# validator = CronValidator
# scheduler = CronScheduler(cronSyntax)
c = CronTab(cronSyntax)
```

```python
lastChecked = datetime.datetime.now()
time.sleep(.6)
now = datetime.datetime.now()


if now.timestamp() + c.previous(default_utc=False) > lastChecked.timestamp():
    print(True)
else:
    print(False)
```
